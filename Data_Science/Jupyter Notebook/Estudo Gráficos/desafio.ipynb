Abaixo temos um dataframe com as seguintes informações:


**Nome:** O nome de cada indivíduo da base


**Idade:** A idade de cada um deles


**Profissão:** A atual ocupação


**Salário:** O salário registrado


**Limite de crédito:** O limite de crédito disponível para cada um dos indivíduos


**Historico_Inadimplencia:** Marco como 0 aqueles que nunca tiveram um histórico de inadimplência e como 1 aqueles que já tiveram em algum momento.


**Estado civil:** O estado civil de cada um deles


**Imovel_Proprio:** Marcado como 0 aqueles que não tem imóvel próprio e como 1 aqueles que tem.

````python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
````

````python

data = {
    'Nome': ['Alice', 'Joao', 'Charlie', 'David', 'Eva', 'Diego', 'Denize', 'Claudio'],
    'Idade': [25, 30, 35, 40, 45, 60, 22, 24],
    'Profissão': ['Engenheiro', 'Médico', 'Professor', 'Advogado', 'Médico','Engenheiro', 'Estudante','Estudante'],
    'Salário': ['4500', '8000', '5000', '10000', '12000','15000', '1200','1500'],
    'Limite_Credito': ['2500', '4000', '4000', '1000', '10000','2000', '500','250'],
    'Historico_Inadimplencia': ['0', '0', '0', '1', '0','1', '0','1'],
    'Estado_Civil': ['Casamento', 'Casamento', 'Solteiro', 'Solteiro', 'Casamento','Solteiro', 'Solteiro','Solteiro'],
    'Imovel_Proprio': ['0', '0', '0', '1', '1','1', '0','0']
}

df = pd.DataFrame(data)
df.head(5)
````

A proposta da atividade é que utilizem os dados do DataFrame Data e escolham apenas 3 gráficos que ajudem vocês a identificar quais fatores são os principais impulsionadores para aqueles que tem limite de crédito alto.

Durante a exploração de vocês podem utilizar mais de 3 gráficos mas deixem na entrega final apenas os 3 mais assertivos que contenham a informação da mensagem que querem passar, fiquem livres para utilizar as bibliotecas que quiserem para os gráficos.

Justifiquem a escolha dos visuais e despertem o lado analítico de vocês com uma análise acerca do descoberto através dos gráficos. Lembrem o que aprendemos na aula, devemos ser assertivos na nossa mensagem.

````python
df['Limite_Credito'] = df['Limite_Credito'].astype(float)
df['Historico_Inadimplencia'] = df['Historico_Inadimplencia'].astype(float)
df['Salário'] = df['Salário'].astype(float)
````

````python
# Nessa etapa criamos a figura da imagem que será gerada, aqui nesses valores numéricos podemos alterar o tamanho da imagem.
plt.figure(figsize=(10, 6))

limite_por_historico = df.groupby('Historico_Inadimplencia')['Limite_Credito'].mean()
plt.bar(limite_por_historico.index, limite_por_historico, color='green')
plt.title('Limite X Histórico', fontsize=16)
plt.xlabel('Histórico de Inadimplencia', fontsize=12)
plt.ylabel('Limite de Crédito', fontsize=12)
plt.xticks(ticks=[0, 1], labels=['Nunca teve nome sujo', 'Já teve nome sujo'], fontsize=10)
plt.grid(False)

plt.show()
````

````python
# Agrupando por profissão para analisar a relação salário-limite de crédito
salario_limite = df.groupby('Profissão')[['Salário', 'Limite_Credito']].mean()

# Plotando o gráfico de barras
plt.figure(figsize=(12, 6))
salario_limite['Salário'].plot(kind='bar', color='blue', alpha=0.6, label='Salário Médio')
salario_limite['Limite_Credito'].plot(kind='bar', color='green', alpha=0.6, label='Limite de Crédito Médio')

# Ajustes do gráfico
plt.title('Relação entre Salário e Limite de Crédito por Profissão', fontsize=16)
plt.ylabel('Valor em R$', fontsize=12)
plt.xlabel('Profissão', fontsize=12)
plt.legend(title='Indicador', bbox_to_anchor=(1.2, 1.2))
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)

plt.show()
````

````python
# Agrupando os dados por Imóvel Próprio e Estado Civil
imovel_limite = df.groupby(['Imovel_Proprio', 'Estado_Civil'])['Limite_Credito'].mean().unstack()

# Plotando gráfico de barras empilhadas
plt.figure(figsize=(10, 6))
imovel_limite.plot(kind='bar', stacked=True, ax=plt.gca(), color=['orange', 'purple'])

# Ajustes do gráfico
plt.title('Relação entre Imóvel Próprio, Estado Civil e Limite de Crédito', fontsize=16)
plt.xlabel('Possui Imóvel Próprio', fontsize=12)
plt.ylabel('Média do Limite de Crédito', fontsize=12)
plt.xticks(ticks=[0, 1], labels=['Não possui (0)', 'Possui (1)'], rotation=0)
plt.legend(title='Estado Civil')
plt.grid(axis='y', linestyle='--', alpha=0.7)

plt.show()
````

## Salário: A base do jogo  
Os gráficos deixam claro que o **salário** é um dos pilares para determinar o limite de crédito. Profissões como médicos e advogados, que possuem uma média salarial mais alta, estão frequentemente associadas a limites de crédito mais elevados. Isso faz sentido: instituições financeiras veem nesses profissionais um menor risco, já que possuem rendas consistentes e previsíveis.

No caso de Eva, com um salário de **R$12.000**, ela garantiu um limite de crédito de **R$10.000**. Já Alice, mesmo sendo engenheira, ainda está construindo sua carreira, e seu salário de **R$4.500** reflete em um limite mais modesto de **R$2.500**.

## Imóveis próprios: O peso da estabilidade  
Ter um **imóvel próprio** também aparece como um diferencial. Pessoas que possuem bens próprios transmitem maior estabilidade financeira, um sinal importante para os bancos. O segundo gráfico mostra que indivíduos com imóvel próprio tendem a ter limites de crédito mais elevados, independentemente de seu estado civil.  

**David**, por exemplo, conseguiu um limite de crédito razoável, mesmo com um histórico de inadimplência, graças à segurança representada por seu imóvel.

## Histórico de inadimplência: A confiança é tudo  
Mas não basta ter um bom salário ou imóvel próprio; o **histórico de inadimplência** pode ser a barreira mais difícil de superar. Clientes que já atrasaram pagamentos enfrentam uma desconfiança natural das instituições financeiras.  

É por isso que **Denize** e **Cláudio**, apesar de jovens e com potencial, possuem limites extremamente baixos. Um histórico limpo é a chave para abrir portas.

## Conexão dos fatores  
Os gráficos e os dados mostram que não há um único fator determinante para um limite de crédito elevado, mas uma **combinação de elementos**:  
- Um bom salário,  
- A posse de bens,  
- E um histórico financeiro sólido.  

Para quem busca melhorar seu limite, a mensagem é clara: investir em estabilidade financeira e evitar inadimplências pode fazer toda a diferença.

## Conclusão  
Os bancos não só avaliam números; eles leem histórias. Cada salário, cada imóvel e cada pagamento feito em dia compõem uma narrativa de confiança e responsabilidade.  

No final, construir um limite de crédito elevado é como contar a melhor versão da sua história financeira.
